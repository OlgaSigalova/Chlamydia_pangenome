```python
from Bio import SeqIO
import sys
import os
import pandas as pd
import numpy as np
from collections import Counter 
import re
import itertools
import fnmatch
```

```python
### Parse original GBK files ###

# input files
path_raw = "/Users/sigalova/Desktop/Chlamydia_pangenome/data/genomes_apr_2019_raw"
gbk_filenames_raw = fnmatch.filter(os.listdir(path_raw), '*.gbff')

# data.frame to write info
genomes_raw_n_cds = pd.DataFrame(columns = ["accession", "n_cds"])
i = 0

# output file
outf = "/Users/sigalova/Desktop/Chlamydia_pangenome/info/genome_n_cds_by_contig_raw_apr2019.csv"
try:
    os.remove(outf)
except:
    pass


for fr in gbk_filenames_raw:
    
    file_raw = path_raw + "/" + fr
    ih_r  = open(file_raw, "r")
    
    assembly_id = "_".join(fr.split("_")[0:2])

    gb_records = list(SeqIO.parse(ih_r, "genbank")) # relevant for gbk files with multiple contigs

    for gb_record in gb_records:

        cds = [x for x in gb_record.features if x.type == "CDS"]
        l = [gb_record.name, len(cds)]        
        print(l)
        genomes_raw_n_cds.loc[i] = np.array(l)
        i += 1
        
genomes_raw_n_cds.to_csv(outf, index = False) 
```
